```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('C:/Users/user/1/Downloads/fifa_ml.csv', encoding='latin-1')
df.head()
```

## Data Cleaning

```python
df.drop(['Unnamed: 0','ID','Photo','Flag','Club Logo','Jersey Number','Loaned From'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
for value in df['Value']:
    for i in range(len(value)):
        if value.isdigit() or value[i] == '.':
            pass
        else:
            value.replace(value[i],"")
```

```python
for i in range(len(df['Value'])):
    new_value = ''
    for char in df['Value'][i]:
        if char.isdigit() or char == '.':
            new_value += char
    df['Value'][i] = new_value

df.head()
```

```python
for i in range(len(df['Wage'])):
    new_value = ''
    for char in df['Wage'][i]:
        if char.isdigit() or char == '.':
            new_value += char
    df['Wage'][i] = new_value

df.head()
```

```python
df['Value'] = pd.to_numeric(df['Value'])
df['Wage'] = pd.to_numeric(df['Wage'])
```

```python
df['Joined'] = pd.to_datetime(df['Joined'])
```

```python
df['Weight'] = df['Weight'].str.rstrip('lbs')
df['Weight'] = pd.to_numeric(df['Weight'])
```

```python
def feet_inches_to_cm(height):
    if pd.isna(height):
        return height
    feet, inches = height.split("'")
    total_inches = int(feet) * 12 + int(inches)
    cm = total_inches * 2.54
    return cm

df['Height'] = df['Height'].apply(feet_inches_to_cm)
```

```python
df['Body Type'].value_counts()
```

```python
value_counts = df['Body Type'].value_counts()
values_to_replace = value_counts[value_counts == 1].index
df['Body Type'].replace(values_to_replace, 'Normal', inplace=True)
df.head()
```

```python
cols = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for column in cols:
    df[column] = df[column].str.replace('\+2', '', regex=True)
    df[column] = df[column].str.replace('\+3', '', regex=True)
    df[column] = df[column].str.replace('\+4', '', regex=True)
    df[column] = df[column].str.replace('\+5', '', regex=True)
    df[column] = pd.to_numeric(df[column])

df.head()
```

```python
for i in range(len(df['Release Clause'])):
    if pd.notna(df['Release Clause'][i]): 
        new_value = ''
        for char in df['Release Clause'][i]:
            if char.isdigit() or char == '.':
                new_value += char
        df['Release Clause'][i] = new_value

df.head()
```

```python
df['Work Rate'].value_counts()
```

```python
## High/High 5 
## High/Medium -- Medium/ High 4

## Medium/Medium --- High/ Low -- Low/High 3
## Medium/Low -- Low/Medium 2
## Low/ Low 1
```

```python
df['Work Rate'] = df['Work Rate'].replace('High/ High', 5)
df['Work Rate'] = df['Work Rate'].replace('High/ Medium', 4)
df['Work Rate'] = df['Work Rate'].replace('Medium/ High', 4)
df['Work Rate'] = df['Work Rate'].replace('Medium/ Medium', 3)
df['Work Rate'] = df['Work Rate'].replace('High/ Low', 3)
df['Work Rate'] = df['Work Rate'].replace('Low/ High', 3)
df['Work Rate'] = df['Work Rate'].replace('Medium/ Low', 2)
df['Work Rate'] = df['Work Rate'].replace('Low/ Medium', 2)
df['Work Rate'] = df['Work Rate'].replace('Low/ Low', 1)
```
